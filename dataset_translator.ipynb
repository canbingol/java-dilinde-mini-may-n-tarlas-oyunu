```python
!pip install googletrans==4.0.0-rc1
import json
from googletrans import Translator

# Dosyanızın yolu
file_path = "instruction-data.json"  # JSON dosyasını buraya yükleyin
output_path = "translated-instruction-data.json"

# JSON dosyasını yükleyin
with open(file_path, "r", encoding="utf-8") as file:
    data = json.load(file)

# Çevirici başlatın
translator = Translator()
translated_data = []

# Çeviri işlemi (ilk 200 örnek için)
for index, entry in enumerate(data, start=1):  # Sadece ilk 200 örneği işliyor
    translated_entry = {
        "talimat": translator.translate(entry["instruction"], src='en', dest='tr').text,
        "girdi": translator.translate(entry["input"], src='en', dest='tr').text if entry["input"] else "",
        "yanıt": translator.translate(entry["output"], src='en', dest='tr').text if entry["output"] else ""
    }
    translated_data.append(translated_entry)

    # Hangi örnekte olduğunu yazdır
    print(f"{index}. örnek çevrildi.")

# Çevrilen veriyi kaydedin
with open(output_path, "w", encoding="utf-8") as file:
    json.dump(translated_data, file, ensure_ascii=False, indent=4)

print("Çeviri tamamlandı ve kaydedildi:", output_path)
```

```python
!pip install googletrans==4.0.0-rc1
import os
from googletrans import Translator

# Çevirici başlat
translator = Translator()

# Dosya yolu
input_file_path = "/content/the-verdict (1).txt"  # Metin dosyanızın adı
output_file_path = "tr_verdict.txt"

# Dosyayı satır satır oku
with open(input_file_path, "r", encoding="utf-8") as file:
    lines = file.readlines()

# Çevir ve yaz
translated_lines = []
for index, line in enumerate(lines, start=1):
    # Boş satırları koru
    if line.strip() == "":
        translated_lines.append("\n")
    else:
        # Satırın Türkçeye çevrilmesi
        translated_line = translator.translate(line, src="en", dest="tr").text
        translated_lines.append(translated_line + "\n")
    print(f"{index}. satır çevrildi.")  # Hangi satırda olduğunu yazdır

# Çevrilen dosyayı kaydet
with open(output_file_path, "w", encoding="utf-8") as file:
    file.writelines(translated_lines)

print(f"Çeviri tamamlandı. Çevrilen dosya: {output_file_path}")
```
